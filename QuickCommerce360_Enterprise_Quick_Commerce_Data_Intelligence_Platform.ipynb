```python
# =====================================================
# PHASE 3 : STAR SCHEMA CREATION
# QuickCommerce360
# =====================================================

import pandas as pd

# =====================================================
# 1. CREATE REVENUE COLUMN
# =====================================================

order_items["revenue"] = (
    order_items["quantity"] *
    order_items["unit_price"]
)

print("Revenue Created")

# =====================================================
# 2. FACT ORDERS
# =====================================================

fact_orders = (
    orders
    .merge(
        order_items,
        on="order_id",
        how="inner"
    )
)

fact_orders = fact_orders[
    [
        "order_id",
        "customer_id",
        "product_id",
        "order_date",
        "quantity",
        "unit_price",
        "revenue",
        "order_total",
        "payment_method"
    ]
]

print("Fact Orders Created")

# =====================================================
# 3. FACT DELIVERY
# =====================================================

fact_delivery = delivery[
    [
        "order_id",
        "delivery_partner_id",
        "delivery_time_minutes",
        "distance_km",
        "delivery_status"
    ]
]

print("Fact Delivery Created")

# =====================================================
# 4. FACT FEEDBACK
# =====================================================

fact_feedback = feedback[
    [
        "feedback_id",
        "order_id",
        "customer_id",
        "rating",
        "sentiment",
        "feedback_category"
    ]
]

print("Fact Feedback Created")

# =====================================================
# 5. FACT MARKETING
# =====================================================

fact_marketing = marketing[
    [
        "campaign_id",
        "date",
        "channel",
        "impressions",
        "clicks",
        "conversions",
        "spend",
        "revenue_generated",
        "roas"
    ]
]

print("Fact Marketing Created")

# =====================================================
# 6. DIM CUSTOMER
# =====================================================

dim_customer = customers[
    [
        "customer_id",
        "customer_segment",
        "area",
        "pincode",
        "registration_date"
    ]
].drop_duplicates()

print("Dim Customer Created")

# =====================================================
# 7. DIM PRODUCT
# =====================================================

dim_product = products[
    [
        "product_id",
        "product_name",
        "category",
        "brand",
        "shelf_life_days"
    ]
].drop_duplicates()

print("Dim Product Created")

# =====================================================
# 8. DIM DATE
# =====================================================

orders["order_date"] = pd.to_datetime(
    orders["order_date"]
)

dim_date = pd.DataFrame()

dim_date["date"] = orders["order_date"].dt.date

dim_date["day"] = orders["order_date"].dt.day

dim_date["month"] = orders["order_date"].dt.month

dim_date["quarter"] = orders["order_date"].dt.quarter

dim_date["year"] = orders["order_date"].dt.year

dim_date["weekday"] = orders["order_date"].dt.day_name()

dim_date = dim_date.drop_duplicates()

print("Dim Date Created")

# =====================================================
# 9. DISPLAY SHAPES
# =====================================================

print("\n")
print("="*50)

print("FACT ORDERS :", fact_orders.shape)

print("FACT DELIVERY :", fact_delivery.shape)

print("FACT FEEDBACK :", fact_feedback.shape)

print("FACT MARKETING :", fact_marketing.shape)

print("DIM CUSTOMER :", dim_customer.shape)

print("DIM PRODUCT :", dim_product.shape)

print("DIM DATE :", dim_date.shape)

print("="*50)

# =====================================================
# 10. SAVE TABLES
# =====================================================

fact_orders.to_csv(
    "fact_orders.csv",
    index=False
)

fact_delivery.to_csv(
    "fact_delivery.csv",
    index=False
)

fact_feedback.to_csv(
    "fact_feedback.csv",
    index=False
)

fact_marketing.to_csv(
    "fact_marketing.csv",
    index=False
)

dim_customer.to_csv(
    "dim_customer.csv",
    index=False
)

dim_product.to_csv(
    "dim_product.csv",
    index=False
)

dim_date.to_csv(
    "dim_date.csv",
    index=False
)

print("\nPhase 3 Completed Successfully")
```

PHASE 4

```python
# =====================================================
# PHASE 4 : DATA QUALITY FRAMEWORK & ETL MONITORING
# QuickCommerce360
# =====================================================

import pandas as pd
from datetime import datetime

print("="*60)
print("PHASE 4 STARTED")
print("="*60)

# =====================================================
# 1. DATA QUALITY REPORT
# =====================================================

quality_report = []

tables = {
    "fact_orders": fact_orders,
    "fact_delivery": fact_delivery,
    "fact_feedback": fact_feedback,
    "fact_marketing": fact_marketing,
    "dim_customer": dim_customer,
    "dim_product": dim_product,
    "dim_date": dim_date
}

for table_name, df in tables.items():

    quality_report.append({
        "table_name": table_name,
        "rows": len(df),
        "columns": len(df.columns),
        "missing_values": df.isnull().sum().sum(),
        "duplicate_rows": df.duplicated().sum()
    })

quality_report_df = pd.DataFrame(quality_report)

print("\nDATA QUALITY REPORT")
display(quality_report_df)

# =====================================================
# 2. REVENUE VALIDATION
# =====================================================

print("\n")
print("="*60)
print("REVENUE VALIDATION")
print("="*60)

total_revenue = round(fact_orders["revenue"].sum(),2)

avg_revenue = round(fact_orders["revenue"].mean(),2)

max_revenue = round(fact_orders["revenue"].max(),2)

print("Total Revenue :", total_revenue)
print("Average Revenue :", avg_revenue)
print("Maximum Revenue :", max_revenue)

# =====================================================
# 3. DELIVERY DATA QUALITY CHECK
# =====================================================

print("\n")
print("="*60)
print("DELIVERY DATA QUALITY")
print("="*60)

negative_delivery_time = fact_delivery[
    fact_delivery["delivery_time_minutes"] < 0
]

print(
    "Negative Delivery Time Records :",
    len(negative_delivery_time)
)

# =====================================================
# 4. INVENTORY DATA QUALITY CHECK
# =====================================================

print("\n")
print("="*60)
print("INVENTORY DATA QUALITY")
print("="*60)

invalid_inventory = inventory[
    inventory["damaged_stock"] >
    inventory["stock_received"]
]

print(
    "Invalid Inventory Records :",
    len(invalid_inventory)
)

# =====================================================
# 5. CUSTOMER DATA QUALITY CHECK
# =====================================================

print("\n")
print("="*60)
print("CUSTOMER DATA QUALITY")
print("="*60)

duplicate_customers = customers.duplicated().sum()

print(
    "Duplicate Customers :",
    duplicate_customers
)

# =====================================================
# 6. PRODUCT DATA QUALITY CHECK
# =====================================================

print("\n")
print("="*60)
print("PRODUCT DATA QUALITY")
print("="*60)

invalid_price = products[
    products["price"] <= 0
]

print(
    "Invalid Product Price Records :",
    len(invalid_price)
)

# =====================================================
# 7. BUSINESS RULE VALIDATION
# =====================================================

print("\n")
print("="*60)
print("BUSINESS RULE VALIDATION")
print("="*60)

invalid_quantity = fact_orders[
    fact_orders["quantity"] <= 0
]

print(
    "Invalid Quantity Records :",
    len(invalid_quantity)
)

invalid_revenue = fact_orders[
    fact_orders["revenue"] <= 0
]

print(
    "Invalid Revenue Records :",
    len(invalid_revenue)
)

# =====================================================
# 8. ETL AUDIT LOG
# =====================================================

etl_audit_log = pd.DataFrame({
    "job_name": [
        "fact_orders_load",
        "fact_delivery_load",
        "fact_feedback_load",
        "fact_marketing_load",
        "dim_customer_load",
        "dim_product_load",
        "dim_date_load"
    ],

    "execution_time": [
        datetime.now()
    ] * 7,

    "status": [
        "SUCCESS"
    ] * 7,

    "rows_processed": [
        len(fact_orders),
        len(fact_delivery),
        len(fact_feedback),
        len(fact_marketing),
        len(dim_customer),
        len(dim_product),
        len(dim_date)
    ]
})

print("\nETL AUDIT LOG")
display(etl_audit_log)

# =====================================================
# 9. SAVE REPORTS
# =====================================================

quality_report_df.to_csv(
    "quality_report.csv",
    index=False
)

etl_audit_log.to_csv(
    "etl_audit_log.csv",
    index=False
)

print("\nReports Saved")

# =====================================================
# 10. PHASE 4 SUMMARY
# =====================================================

phase4_summary = pd.DataFrame({
    "Metric":[
        "Total Revenue",
        "Average Revenue",
        "Max Revenue",
        "Negative Delivery Records",
        "Invalid Inventory Records"
    ],
    "Value":[
        total_revenue,
        avg_revenue,
        max_revenue,
        len(negative_delivery_time),
        len(invalid_inventory)
    ]
})

print("\nPHASE 4 SUMMARY")
display(phase4_summary)

print("\n")
print("="*60)
print("PHASE 4 COMPLETED SUCCESSFULLY")
print("="*60)
```

PHASE 5

```python
# =====================================================
# PHASE 5 : FEATURE ENGINEERING & BUSINESS KPIs
# QuickCommerce360
# =====================================================

import pandas as pd
import numpy as np

print("="*60)
print("PHASE 5 STARTED")
print("="*60)

# =====================================================
# 1. CUSTOMER KPIs
# =====================================================

customer_kpis = (
    fact_orders
    .groupby("customer_id")
    .agg(
        total_orders=("order_id","count"),
        total_revenue=("revenue","sum"),
        avg_order_value=("revenue","mean")
    )
    .reset_index()
)

customer_kpis["customer_lifetime_value"] = (
    customer_kpis["total_revenue"]
)

customer_kpis["purchase_frequency"] = (
    customer_kpis["total_orders"]
)

print("Customer KPIs Created")

# =====================================================
# 2. CUSTOMER SEGMENTATION
# =====================================================

customer_kpis["customer_segment"] = pd.qcut(
    customer_kpis["customer_lifetime_value"],
    q=3,
    labels=["Bronze","Silver","Gold"]
)

print("Customer Segmentation Created")

# =====================================================
# 3. PRODUCT KPIs
# =====================================================

product_kpis = (
    fact_orders
    .groupby("product_id")
    .agg(
        total_quantity=("quantity","sum"),
        total_revenue=("revenue","sum")
    )
    .reset_index()
)

product_kpis = (
    product_kpis
    .merge(
        dim_product,
        on="product_id",
        how="left"
    )
)

print("Product KPIs Created")

# =====================================================
# 4. CATEGORY KPIs
# =====================================================

category_kpis = (
    product_kpis
    .groupby("category")
    .agg(
        category_revenue=("total_revenue","sum"),
        quantity_sold=("total_quantity","sum")
    )
    .reset_index()
)

print("Category KPIs Created")

# =====================================================
# 5. DELIVERY KPIs
# =====================================================

delivery_kpis = pd.DataFrame()

delivery_kpis["avg_delivery_time"] = [
    fact_delivery["delivery_time_minutes"].mean()
]

delivery_kpis["avg_distance"] = [
    fact_delivery["distance_km"].mean()
]

delivery_kpis["on_time_deliveries"] = [
    len(
        fact_delivery[
            fact_delivery["delivery_status"]=="On Time"
        ]
    )
]

delivery_kpis["late_deliveries"] = [
    len(
        fact_delivery[
            fact_delivery["delivery_status"]!="On Time"
        ]
    )
]

delivery_kpis["delivery_sla_percent"] = [
    (
        len(
            fact_delivery[
                fact_delivery["delivery_status"]=="On Time"
            ]
        )
        /
        len(fact_delivery)
    ) * 100
]

print("Delivery KPIs Created")

# =====================================================
# 6. FEEDBACK KPIs
# =====================================================

feedback_kpis = (
    fact_feedback
    .groupby("sentiment")
    .size()
    .reset_index(name="count")
)

avg_rating = fact_feedback["rating"].mean()

print("Feedback KPIs Created")

# =====================================================
# 7. MARKETING KPIs
# =====================================================

marketing_kpis = marketing.copy()

marketing_kpis["conversion_rate"] = (
    marketing_kpis["conversions"]
    /
    marketing_kpis["clicks"]
) * 100

marketing_kpis["cost_per_conversion"] = (
    marketing_kpis["spend"]
    /
    marketing_kpis["conversions"]
)

print("Marketing KPIs Created")

# =====================================================
# 8. RFM ANALYSIS
# =====================================================

fact_orders["order_date"] = pd.to_datetime(
    fact_orders["order_date"]
)

snapshot_date = (
    fact_orders["order_date"].max()
    + pd.Timedelta(days=1)
)

rfm = (
    fact_orders
    .groupby("customer_id")
    .agg({
        "order_date":
            lambda x: (
                snapshot_date - x.max()
            ).days,
        "order_id":"count",
        "revenue":"sum"
    })
)

rfm.columns = [
    "Recency",
    "Frequency",
    "Monetary"
]

rfm = rfm.reset_index()

print("RFM Analysis Created")

# =====================================================
# 9. TOP CUSTOMERS
# =====================================================

top_customers = (
    customer_kpis
    .sort_values(
        "total_revenue",
        ascending=False
    )
    .head(10)
)

print("Top Customers Created")

# =====================================================
# 10. SAVE FILES
# =====================================================

customer_kpis.to_csv(
    "customer_kpis.csv",
    index=False
)

product_kpis.to_csv(
    "product_kpis.csv",
    index=False
)

category_kpis.to_csv(
    "category_kpis.csv",
    index=False
)

delivery_kpis.to_csv(
    "delivery_kpis.csv",
    index=False
)

feedback_kpis.to_csv(
    "feedback_kpis.csv",
    index=False
)

marketing_kpis.to_csv(
    "marketing_kpis.csv",
    index=False
)

rfm.to_csv(
    "rfm_segmentation.csv",
    index=False
)

top_customers.to_csv(
    "top_customers.csv",
    index=False
)

# =====================================================
# 11. SUMMARY
# =====================================================

print("\n")
print("="*60)
print("PHASE 5 SUMMARY")
print("="*60)

print("Customer KPI Records :", len(customer_kpis))
print("Product KPI Records :", len(product_kpis))
print("Category KPI Records :", len(category_kpis))
print("Marketing KPI Records :", len(marketing_kpis))
print("RFM Records :", len(rfm))

print("\nPhase 5 Completed Successfully")
```
